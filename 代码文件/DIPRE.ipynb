```python
    import requests        #导入requests包
    from bs4 import BeautifulSoup
    url = 'http://www.ireader.com/index.php?ca=search.index&keyword=%E6%9C%B1%E8%87%AA%E6%B8%85'
    strhtml = requests.get(url)        #Get方式获取网页数据
    strhtml.encoding='utf-8'
    soup = BeautifulSoup(strhtml.text) #将得到的网页利用BeautifulSoup进行剖析
    #print(strhtml.text)
    soup.body.a #通过点取属性的方式只能获得当前名字的第一个tag:
    #print(soup.select('.sResult'))
```

```python
print(soup.prettify())
```

```python
print(soup.title)
print(soup.title.name) 
print(soup.title.string)
print(soup.title.parent.name)
print(soup.p)
soup.p['class']
soup.a
soup.find_all('a') #如果想要得到所有的<a>标签,或是通过名字得到比一个tag更多的内容的时候,就需要用到 Searching the tree 中描述的方法,比如: find_all()
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
soup0 = BeautifulSoup('<b class="boldest res" color="red">Extremely bold</b>')
tag = soup0.b
tag.name
tag.name = "black" #修改标签名
tag['class'] #获取属性
tag.attrs #获取所有属性
tag['id'] = 1 #修改属性
del tag['class'] #删除属性
tag.string.replace_with('no longer') #tag中包含的字符串不能编辑,但是可以被替换成其它的字符串,用 replace_with() 方法
tag.string #输出标签中的字符串
```

```python
url = 'http://www.ireader.com/index.php?ca=search.index&keyword=%E6%9C%B1%E8%87%AA%E6%B8%85'
strhtml = requests.get(url)        #Get方式获取网页数据
strhtml.encoding='utf-8'
soup = BeautifulSoup(strhtml.text) #将得到的网页利用BeautifulSoup进行剖析
head_tag = soup.head
head_tag.contents
head_tag.contents[3].name #标签的 .contents 属性可以将标签的子节点以列表的方式输出:
```
