```python
import re
match = re.search(r'[1-9]\d{5}','BIT 100081')
if match:
    print(match.group(0))
```

```python
import re
match = re.match(r'[1-9]\d{5}','1004081')
if match:
    print(match.group(0))  #NULL
```

```python
import re
ls = re.findall(r'[1-9]\d{5}', 'BIT100081 TSU100084')
print(ls)
```

```python
import re
for m in re.finditer(r'[1-9]\d{5}', 'BIT100081 TSU100089 AA135245'):
    if m:
        print(m) #100081 100084
```

```python
import requests        #导入requests包
from bs4 import BeautifulSoup
from lxml import etree
url = 'https://www.rainydaypaperback.com/products/category/3219/~/~/gardening'
strhtml = requests.get(url)        #Get方式获取网页数据
strhtml.encoding='utf-8'
soup = BeautifulSoup(strhtml.text) 
print(soup.body.text.replace("\n",""))
```

```python
from lxml import etree
 
html = strhtml.text
_element = etree.HTML(html)
text = _element.xpath('//body//text()')
print(text)
```
