```python
import requests
from bs4 import BeautifulSoup

def get_stars(soup):
    nostar = soup.find_all('span',{'class','no'})
    return 5-nostar.__len__() # 评分为5-空星星数组长度

def get_bookname_lists(root , url , page):
    booklist = []
    while page:
        path = root + url
        html = requests.get(path)
        soup = BeautifulSoup(html.text)
        body = soup.find('div','yd-book-content yd-book-content-standalone yd-store-content-container clearfix')
        for item in body.find_all('div','yd-book-item yd-book-item-pull-left'):
            name = item.find('h2').text
            author = item.find('dd').text
            link = item.find('a')['href']
            price = item.find('b').text.strip() # 去除左右两边的空格
            introduction = item.find('div',{'class','summery'}).find('p').text
            stars = get_stars(item.find('div',{'class','rate w-star w-star1'}))
            booklist.append([name,author,price,stars])
        url = soup.find(text = '下一页').parent['href']
        page = page-1
    return booklist
        
root = "http://yuedu.163.com"
url = "/book/category/category/0_0_1/p1/s20"
print(get_bookname_lists(root,url,10))
```
